# LẤY MODEL VỀ

```python
import torch
print("CUDA available:", torch.cuda.is_available())
```

```python
from transformers import pipeline
fiction_categories = ["Fiction", "Nonfiction", "Children's Fiction", "Children's Nonfiction"]
pipe =  pipeline("zero-shot-classification",
                 model="facebook/bart-large-mnli",
                 device=0)
```

```python
sequence1 = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
pipe(sequence1, fiction_categories)
```

```python
sequence2 = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[0]
```

```python
pipe(sequence2, fiction_categories)
```

```python
import pandas as pd
def test_predict_categories(item_sum):
    actual_cats = []
    predicted_cats = []
    fiction_texts = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[:item_sum].tolist()
    nonfiction_texts = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[:item_sum].tolist()
    print("ĐANG XỬ LÝ FICTION ....")
    fiction_results = pipe(fiction_texts, candidate_labels=fiction_categories, batch_size=16)
    print("ĐÃ XỬ LÝ XONG FICTION .....")
    print("ĐANG XỬ LÝ NONFICTION")
    nonfiction_results = pipe(nonfiction_texts, candidate_labels=fiction_categories, batch_size=16)
    print("ĐÃ XỬ LÝ XONG NONFICTION")
    predicted_cats = [res['labels'][0] for res in fiction_results] + [res['labels'][0] for res in nonfiction_results]
    actual_cats = ["Fiction"] * 300 + ["Nonfiction"] * 300
    predictions_df = pd.DataFrame({
        "actual_categories": actual_cats,
        "predicted_categories": predicted_cats
    })
    return predictions_df
```

```python
predictions_df = test_predict_categories(300)
predictions_df
```

```python
import numpy as np
predictions_df["correct_predicted"] = (
    np.where(predictions_df["actual_categories"] == predictions_df["predicted_categories"], 1, 0)
)
```

```python
predictions_df["correct_predicted"].sum()/len(predictions_df)
```

## BÂY GIỜ ĐI XỬ LÝ CÁC CATEGORIES CHƯA ĐƯỢC MAP, Ở SIMPLE_CATEGORIES ĐANG LÀ NaN

```python
print(books['categories'].isna().sum())
```

```python
books.dropna(subset=['categories'], inplace=True)
```

```python
print(books['categories'].isna().sum())
```

```python
missing_cats = books.loc[books["simple_categories"].isna(), ["isbn13", "description", "categories"]].reset_index(drop=True)
```

```python
def predict_categories(missing_cats):
    actual_cats = []
    isbns = []
    predicted_cats = []
    for i in range(len(missing_cats)):
        actual_cats += [missing_cats['categories'][i]]
        isbns += [missing_cats["isbn13"][i]]
    sequences = missing_cats["description"].astype(str).tolist()
    predict_result = pipe(sequences, candidate_labels=fiction_categories, batch_size=16)
    predicted_results = [res['labels'][0] for res in predict_result]
    missing_predicted_df = pd.DataFrame({"isbn13": isbns, "predicted_categories": predicted_results, "actual_categories": actual_cats})
    return missing_predicted_df
```

```python
missing_predicted_df = predict_categories(missing_cats)
missing_predicted_df.head(50)
```

## HƠN 1400 ROWS NÊN SẼ MẤT VÀI PHÚT

```python
missing_predicted_df
```

```python
missing_predicted_df = missing_predicted_df.drop(columns=["actual_categories"])
books = pd.merge(books, missing_predicted_df, on="isbn13", how='left')
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"],
                                      books["simple_categories"])
books
```
